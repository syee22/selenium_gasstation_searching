## 외부 파일 읽어오기
 - 이름이 비슷한 같은 형식의 파일이 여러개 있을 때 한번에 읽어오는 방법
 - 관련 패키지 : glob
    - 파일 경로 및 이름을 모아서 리스트에 저장
    - * 문자 사용 가능
    - ex. 주유소*.xlsx (주유소로 시작하는 모든 xlsx 파일)

```python
from glob import glob
```

```python
f = glob('./crawl_data/지역_위치별*.xls')
```

```python
f
```

f= pd.read_excel(f)

```python
tmp_raw =[]

for f_name in f :
    tmp = pd.read_excel(f_name,header=2)
    tmp_raw.append(tmp)
```

```python
tmp_raw
```

```python
station_raw = pd.concat(tmp_raw)
```

```python
station_raw.shape
station_raw.head()
```

```python
station = pd.DataFrame({'주유소':station_raw['상호'],
                        '주소': station_raw['주소'],
                        '가격': station_raw['휘발유'],
                        '셀프': station_raw['셀프여부'],
                        '상표': station_raw['상표']})
```

```python
station.head()
```

```python
# 구별 가공데이터 생성하기 위해 변수 추가
station['구']=[addr.split()[1] for addr in station['주소']]
```

```python
station.head()
```

```python
station_fin = station[station['가격']!='-']
```

```python
# 인덱스 설정
station_fin.reset_index(inplace=True)
```

```python
station_fin.tail()
```

```python
# 가격이 문자열형이어서 실수형으로 변경
station_fin['가격'] = [float(price) for price in station_fin['가격']]
```

```python
station_fin.info()
```

### 서울시 구별 주유가격 단계 구분도 

### 파이썬 단계구분도
- folium 패키지의 choropleth() 사용
1. folium.Map(location=[중심위치 위경도], zoom_start=초기 zoom 비율, tiles=지도 형태)함수를 이용해 기본 지도 생성
2. 기본 지도 위에 단계 구분도를 추가
    - choropleth(geo_data=경계위경도가 포함된 json파일,data, columns,fill_color,key_on=)


```python
import json
import folium
```

```python
geo_path='../ch05__data/data/02. skorea_municipalities_geo_simple.json'
```

```python
# 구별 주유가격 평균표 생성 
gu_data = pd.pivot_table(station_fin,index=['구'],
                         values=['가격'],
                         aggfunc=np.mean)
gu_data
```

```python
# geo_data 불러오기
geo_str = json.load(open(geo_path, encoding='utf-8'))
geo_str
```

```python
map = folium.Map(location=[37.5502,126.982],
                zoom_start=10.5,
                tiles='Stamen Toner')
```

```python
map.choropleth(geo_data=geo_str,
              data=gu_data,
              columns=[gu_data.index,'가격'],
               fill_color='YlGnBu',
              key_on='feature.id')
```
